## 1. 중복 삭제 후 pivot table 생성

```python
df.drop_duplicates(subset=["Part Number", "CRET_TIME"], inplace=True)
pivot_table = df.pivot(index="Part Number", columns="CRET_TIME", values="D일 투입예정 수량(D일계획)")
pivot_table
```

```python
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Train/Test split

```python
X = pivot_table.iloc[:, :-10]
```

```python
y = pivot_table.iloc[:, -10:]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    train_size = 0.8,
                                                    random_state=100)
X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

standardScaler = StandardScaler().fit(X_train)
X_train = standardScaler.transform(X_train)
X_test = standardScaler.transform(X_test)
```

```python
# from sklearn.preprocessing import MinMaxScaler

# minmaxscaler = MinMaxScaler().fit(X_train)
# X_train = minmaxscaler.transform(X_train)
# X_test = minmaxscaler.transform(X_test)
```

## 3. DataLoader 구축

```python
from torch.utils.data import TensorDataset, DataLoader

X_train = np.array(X_train).astype(np.float32)
y_train = np.array(y_train).astype(np.float32)
X_train_norm = ((X_train-np.mean(X_train))/np.std(X_train))
X_train_norm = torch.from_numpy(X_train_norm)
y_train = torch.from_numpy(y_train)

train_ds = TensorDataset(X_train_norm, y_train)

batch_size = 1
train_dl = DataLoader(train_ds,
                      batch_size,
                      shuffle=True)
```

```python
train_dl
```

## 4. Model Build

```python
import torch.nn as nn

# 모델 구현
model = nn.Sequential(
    nn.Linear(143, 10, bias=True), # input layer
    nn.ReLU(), # activation function
    nn.Linear(10, 10, bias=True), # hidden layer
    nn.ReLU(),
    nn.Linear(10, 10, bias=True), # output layer
    nn.ELU(),
    nn.Linear(10, 10, bias=True) # output layer
    # nn.Sigmoid() # activation function
    )
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
```

## 5. Training

```python
for epoch in range(200):
  for x_batch, y_batch in train_dl: # enumarate 안 쓴 경우
    pred = model(x_batch)

    loss = criterion(pred, y_batch)
    loss.backward()

    optimizer.step()
    optimizer.zero_grad()

  if epoch % 10 == 0:
    print(f'Epoch {epoch}: Loss {loss.item():.4f}')
```

## 6. Prediction

```python
X_test = np.array(X_test).astype(np.float32)
X_test_norm = (X_test - np.mean(X_train))/np.std(X_train)
X_test_norm = torch.from_numpy(X_test_norm)

y_pred = model(X_test_norm).detach().numpy() # no_grad를 한다.
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def evaluate(y_test, y_pred):
  print('* MAE:', mean_absolute_error(y_test, y_pred))
  print('* RMSE:', np.sqrt(mean_absolute_error(y_test, y_pred)))
  print('* R_square:', r2_score(y_test, y_pred))
```

```python
evaluate(y_test, y_pred)
```
