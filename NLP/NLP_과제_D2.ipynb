```python
# key와 value를 딕셔너리로 묶음
data = dict(zip(df['title'],df.index))
data
```

```python
find_title = df['title'].head(1)[0]
```

```python
def recommend(t, cosine_sim=cos_sim): # t에는 제목 전달
    idx = data[t] #
    sim_scores=list(enumerate(cos_sim[idx])) # enumerate : 인덱스번호와 유사도가 쌍으로 나옴
    ss=sorted(sim_scores, key=lambda x: x[1], reverse=True)
    ss=ss[1:11]
    res = [i[0] for i in ss] # index번호와 유사도 값
    return df['title'].iloc[res]
```

```python
recommend(find_title)
```
