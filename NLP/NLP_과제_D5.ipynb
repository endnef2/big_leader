```python
# 단어에 대한 전처리(모델 성능 개선)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from string import punctuation
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.models import Sequential
nltk.download('stopwords')
nltk.download('punkt')
```

```python
def remove_stopwords(text_list):
    stop_words = set(stopwords.words('english'))
    filtered_texts = []
    tok=Tokenizer()
    tok.fit_on_texts(x_train)
    for text in text_list:
        words = word_tokenize(text)  # 텍스트를 단어 단위로 토큰화
        filtered_text = [word for word in words if len(word) > 2 and word.lower() not in stop_words and not any(c in word for c in "'s" + punctuation)]
        filtered_texts.append(' '.join(filtered_text))
    return filtered_texts
```

```python
x_train = remove_stopwords(x_train)
x_test = remove_stopwords(x_test)
```

```python
x_test
```

```python
tok=Tokenizer()
tok.fit_on_texts(x_train) # xtrain을 코퍼스로 하여 fitting(단어별로 index 부여됨), 원래는 전체에 대해 하고 이후에 나누기
x_train_enc = tok.texts_to_sequences(x_train)
```

```python
# 최대길이 찾기
x_train_enc
# map(len, data) , # map(적용함수, data)
sum(map(len, x_train_enc))/len(x_train_enc) # 메일의 평균 단어 갯수
max(map(len, x_train_enc))
min(map(len, x_train_enc))
```

```python
max_len = max(len(s) for s in x_train_enc)
```

```python
max_len
```

```python
min(len(s) for s in x_train_enc)
```

```python
x_train_padded=pad_sequences(x_train_enc, maxlen=max_len)
```

```python
x_train_padded.shape
```

```python
# vocab_size = len(w2i) # 1번부터 단어의 번호 부여
vocab_size = len(w2i)+1 # 0 패딩을 했으므로 1을 더하여 단어사전크기를 설정
```

```python
# RNN
from tensorflow.keras.layers import SimpleRNN, Embedding, Dense, LSTM
from tensorflow.keras.models import Sequential
```

```python
model=Sequential()
model.add(Embedding(vocab_size, 32))
model.add(SimpleRNN(100))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(x_train_padded, y_train, epochs=5, batch_size=32, validation_split=0.2)
```

```python
# 모델의 구조와 weight가 저장됨
```

```python
x_test_enc=tok.texts_to_sequences(x_test)
```

```python
x_test_padded = pad_sequences(x_test_enc, maxlen=max_len)
```

```python
model.predict(x_test_padded)
```

```python
model.evaluate(x_test_padded, y_test)
```
